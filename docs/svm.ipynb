```python
from autograd import numpy as np
from sparse_vi.svm import SVM
import matplotlib
%matplotlib inline
```

```python
rs = np.random.RandomState(seed=1234)
model = SVM(μ0=0., Vμ=.5, φ0=.9, Vφ=.5**2, γ0=np.zeros(2), Ω0=0.1*np.eye(2),
            β0=0., Vβ=.05, νσ2=10, Sσ2=.2**2*(10-1),
            νΩ=10, SΩ=(10+3)*np.diag([.1**2, .25**2]))
θ0 = {'μ': 0.01, 'φ':  0.98, 'β': 0.51, 'σ2': 0.05,
      'Ω': np.array([[0.015, 0.018], [0.018,  0.043]])}
data = model.simulate(T=1000, **θ0, rs=rs)
```

## Data

```python
data.plot_data()
```

```python
data.plot_moving_params()
```

```python
data.plot_state()
```
